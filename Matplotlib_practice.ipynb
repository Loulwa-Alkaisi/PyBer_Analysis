```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
#Line Chart Using the MATLAB Method
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
#Line Chart Using the Object-Oriented Interface method.
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#Another way the ax.plt() can be written.
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label = "Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
#Add a gird.
plt.grid()
# Add the legend.
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label= "Boston")
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_ylim(0, 45)
ax.set_title("PyBer Fare by Month")
ax.grid()
ax.legend()
```

```python
#Scatter plots.
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis, c="red", label="Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0,45)
plt.gca().invert_yaxis()
plt.legend()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
#list comprehension technique (could be used instead of the for loop above.)
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
#Create a Scatter Plot Using the Object-Oriented Interface
fig, ax = plt.subplots()
plt.scatter(y_axis, x_axis, c ='skyblue', alpha = 0.5, edgecolor= "k", linewidth=2, s=[i*5 for i in y_axis], label = "Boston")
ax.set_title("PyBer Fair by Month")
ax.set_xlim(0,45)
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
plt.gca().invert_yaxis()
ax.legend()
```

```python
#Create a pie chart.
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
        explode=explode_values,
        colors = colors,
        labels=x_axis,
        autopct='%1.1f%%')
plt.show()
```

```python
#Create a Pie Chart Using the Object-Oriented Interface.
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

```python
colors = ["cyan", "darkturquoise","olivedrab", "greenyellow", "yellow", "khaki", "orangered", "firebrick", "crimson", "hotpink", "violet", "darkviolet"]
explode_values = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]
plt.subplots(figsize=(8,8))
plt.pie(y_axis,
       explode = explode_values,
       colors = colors,
       shadow = "false",
       counterclock = False,
       labels = x_axis,
       autopct = ('%1.1f%%'))

plt.show()
```
